```python
# cargamos las librerías que usaremos

from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from tensorflow import keras
from keras import Sequential
import tensorflow as tf
import random
import shutil
import os
```

```python
# Vamos a crear nuestro train DataSet

# Primero creamos una carpeta donde se almacenarán las imágenes de prueba
path = "/data/test" # reemplazar por la carpeta donde se tenga el set de imágenes

try:
  os.mkdir(path)
except:
  print('La carpeta ya existe')


# Creamos las carpetas de nuestro data test y luego de manera aleatoria movemos el 20% de los archivos

content_data = '/data/FRUIT-16K'  # reemplazar por la carpeta donde se tenga el set de imágenes
content_test = '/data/test'       # reemplazar por la carpeta donde se quiera guardar el set de imágenes

content_folders = os.listdir(content_data)
folder_test = ''

for i in range(len(content_folders)):
  folder_test = content_test + '/' +  content_folders[i]
  files = os.listdir(content_data + '/' + content_folders[i])

  try:
    os.mkdir(folder_test)
  except:
    print('La carpeta ya existe')

  size_list = len(files)
  size_test = size_list*0.2

  for j in range(int(size_test)):
    choose_file = random.choice(files)
    path_file = content_data + '/' + content_folders[i] + '/' + choose_file
    new_path_file = folder_test + '/'

    try:
      shutil.move(path_file, new_path_file)
    except:
      print('no fue posible')

    files = os.listdir(content_data + '/' + content_folders[i])
```

Lo inicial es crear una forma de construir un dataset de prueba y separar los archivos en carpetas del sistema, para eso con funciones de diferentes librerías se examina las carpetas del dataset y se se recorre cada carpeta para:


1.   Crear una carpeta "espejo" en una carpeta nombrada test
2.   Buscar todos los archivos de cada categoría
3.   Mover de manera aleatoria el 20% de cada carpeta
